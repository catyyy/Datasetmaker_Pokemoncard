```python
import os
import cv2
import json
import requests
!pip install ultralytics
from ultralytics import YOLO
#import the model
model = YOLO("./best.onnx")
class_names = model.names
```

```python
def extract_named_boxes(image, results, class_name='Name'):
    class_names = results.names
    for box in results.boxes:
        class_id = int(box.cls.item())
        label = class_names[class_id]
        if label == class_name:
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            cropped_img = image[y1:y2, x1:x2]
            return cropped_img

def extract_pic_boxes(results, class_name='Pic'):
    class_names = results.names
    for box in results.boxes:
        class_id = int(box.cls.item())
        label = class_names[class_id]
        if label == class_name:
          x1, y1, x2, y2 = map(float, box.xyxy[0].tolist())
          return x1, y1, x2, y2

def display_image_with_matplotlib(image):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()

def turn_pic_to_name(image):
  endpoint = "https://kprzyjocr.cognitiveservices.azure.com/"
  subscription_key = "fd4e6b8a6b0d4bb68312e0ea7fe06f07"
  #make sure size > 50 pixel
  original_height, original_width = image.shape[:2]
  scale_factor = max(100 / original_width, 100 / original_height)
  new_width = int(original_width * scale_factor)
  new_height = int(original_height * scale_factor)
  resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
  border_size = 12
  white = [255, 255, 255]
  padded_image = cv2.copyMakeBorder(image, border_size, border_size, border_size, border_size, cv2.BORDER_CONSTANT, value=white)
  display_image_with_matplotlib(padded_image)

  _, image_encoded = cv2.imencode('.jpg', padded_image)
  image_bytes = image_encoded.tobytes()
  ocr_url = f"{endpoint}/vision/v3.2/ocr"
  headers = {
      'Ocp-Apim-Subscription-Key': subscription_key,
      'Content-Type': 'application/octet-stream'
  }
  params = {
      'language': 'ja'
  }
  response = requests.post(ocr_url, headers=headers, params=params, data=image_bytes)

  try:
      response.raise_for_status()
      result = response.json()
      #print(result)
      for region in result.get('regions', []):
          for line in region.get('lines', []):
              line_text = ''.join([word['text'] for word in line.get('words', [])])
              return line_text

  except requests.exceptions.HTTPError as e:
      print(f"HTTP error occurred: {e}")
      print(f"Response content: {response.text}")

def find_or_add_name(name_list, name):
    if name in name_list:
        return name_list.index(name)
    else:
        name_list.append(name)
        return name_list.index(name)

labels = []
for filename in os.listdir('./images'):
  if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
    name_without_ext = os.path.splitext(filename)[0]
    file_path = os.path.join('./images', filename)
  results = model(file_path)  # predict on an image
  import matplotlib.pyplot as plt
  for result in results:
    Name_Pic = extract_named_boxes(result.orig_img, result, class_name='Name')
    display_image_with_matplotlib(Name_Pic)
    Name_txt = turn_pic_to_name(Name_Pic)
    print(Name_txt)
    label_no = find_or_add_name(labels, Name_txt)
    new_txt_filename = os.path.join('./labels', f'{name_without_ext}.txt')
    pic_box = extract_pic_boxes(result, class_name='Pic')
    img_height, img_width = result.orig_img.shape[:2]
    x_min, y_min, x_max, y_max = pic_box
    box_width = x_max - x_min
    box_height = y_max - y_min
    x_center = (x_min + box_width / 2) / img_width
    y_center = (y_min + box_height / 2) / img_height
    width_norm = box_width / img_width
    height_norm = box_height / img_height
    with open(new_txt_filename, 'w', encoding='utf-8') as new_file:
      new_file.write(f'{label_no} {x_center:.6f} {y_center:.6f} {width_norm:.6f} {height_norm:.6f}')

with open('./labels/labels.txt', 'w', encoding='utf-8') as file:
  for item in labels:
    file.write(f'{item}\n')
```
